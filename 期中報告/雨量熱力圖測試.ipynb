```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 從 CSV 檔案讀取資料
file_path = 'C:\\Users\\Jefferson\\Desktop\\智慧城市\\期中報告資料\\整理資料\\2023臺北市水利處歷史日雨量-單日最大經緯度UTF8.csv'
data = pd.read_csv(file_path)

# 繪製熱力圖
plt.figure(figsize=(10, 8))
heatmap = sns.scatterplot(
    x='經度', y='緯度', size='單日雨量', data=data, sizes=(20, 200),
    legend=False, palette='viridis', alpha=0.8
)

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']  # 設置字型為微軟正黑體或其他中文字型

# 添加標籤
plt.xlabel('經度')
plt.ylabel('緯度')
plt.title('2023年單日雨量熱力圖')
plt.show()
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# 從 CSV 檔案讀取資料
file_path = 'C:\\Users\\Jefferson\\Desktop\\智慧城市\\期中報告資料\\整理資料\\2023臺北市水利處歷史日雨量-單日最大經緯度UTF8.csv'
data = pd.read_csv(file_path)

# 創建 Folium 地圖
m = folium.Map(location=[25.033, 121.565], zoom_start=12)  # 台北市的經緯度

# 將單日雨量資料轉換為 [緯度, 經度, 權重] 的列表
heat_data = [[row['緯度'], row['經度'], row['單日雨量']] for index, row in data.iterrows()]

# 在地圖上添加熱力圖層
HeatMap(heat_data, radius=15).add_to(m)

# 顯示地圖
m
```
